```python
state_cols = {
    "NSW": "Estimated Resident Persons Population NSW",
    "Victoria": "Estimated Resident Persons Population  Victoria ",
    "Queensland": "Estimated Resident Persons Population Queensland ",
    "South Australia": "Estimated Resident Persons Population  South Australia ",
    "Western Australia": "Estimated Resident Persons Population  Western Australia ",
    "Tasmania": "Estimated Resident Persons Population  Tasmania ",
    "Northern Territory": "Estimated Resident Persons Population  Northern Territory ",
    "ACT": "Estimated Resident Persons Population Australian Capital Territory "
}

aus_col = "Estimated Resident Persons Population Australia "
```

```python
# --- compute growth ---
growth_df = df4[list(state_cols.values())].pct_change() * 100
growth_df.columns = list(state_cols.keys())

growth_df["Year"] = df4.index
growth_df = growth_df.set_index("Year")
growth_df = growth_df.dropna(how="all")

# --- plot ---
plt.figure(figsize=(12,8))

plt.imshow(growth_df.T.fillna(0), aspect="auto")
plt.colorbar(label="YoY Growth %")

plt.yticks(range(len(growth_df.columns)), growth_df.columns)

step = max(len(growth_df.index)//10, 1)
plt.xticks(
    range(0, len(growth_df.index), step),
    growth_df.index[::step],
    rotation=45
)

plt.title("State Population Growth Heatmap")
plt.xlabel("Year")
plt.ylabel("State")

plt.tight_layout()
plt.show()
```

```python
df4.index = pd.to_datetime(df4.index)

# yearly average population
df_yearly = df4.resample("Y").mean()

# optional: convert index to just year
df_yearly.index = df_yearly.index.year
```

```python
growth_df = df_yearly[list(state_cols.values())].pct_change() * 100
growth_df.columns = list(state_cols.keys())
growth_df = growth_df.dropna(how="all")
```

```python
plt.figure(figsize=(12,6))

# clip extreme outliers for better contrast
heat_data = growth_df.T.clip(lower=-1, upper=3)

plt.imshow(heat_data, aspect="auto")
plt.colorbar(label="YoY Growth %")

plt.yticks(range(len(heat_data.index)), heat_data.index)

step = max(len(heat_data.columns)//12, 1)
plt.xticks(
    range(0, len(heat_data.columns), step),
    heat_data.columns[::step],
    rotation=45
)

plt.title("State Population Growth Heatmap (Yearly)")
plt.xlabel("Year")
plt.ylabel("State")

plt.tight_layout()
plt.show()
```

```python
# Australia column name
aus_col = 'Estimated Resident Persons Population Australia '

# compute YoY
aus_yoy = df_yearly[aus_col].pct_change() * 100

plt.figure(figsize=(10,5))
plt.plot(df_yearly.index, aus_yoy, linewidth=2)

plt.axhline(0, linestyle="--")
plt.title("Australia Population YoY Growth")
plt.xlabel("Year")
plt.ylabel("Growth %")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# compute shares
share_df = pd.DataFrame(index=df_yearly.index)

for state, col in state_cols.items():
    share_df[state] = df_yearly[col] / df_yearly[aus_col] * 100

plt.figure(figsize=(12,6))

plt.stackplot(
    share_df.index,
    [share_df[state] for state in share_df.columns],
    labels=share_df.columns
)

plt.title("State Share of National Population")
plt.xlabel("Year")
plt.ylabel("Share (%)")
plt.legend(loc="upper left", bbox_to_anchor=(1,1))

plt.tight_layout()
plt.show()
```

```python
cagr = {}

years = df_yearly.index[-1] - df_yearly.index[0]

for state, col in state_cols.items():
    start = df_yearly[col].iloc[0]
    end = df_yearly[col].iloc[-1]
    cagr[state] = ((end/start)**(1/years) - 1) * 100

cagr_series = pd.Series(cagr).sort_values(ascending=False)

plt.figure(figsize=(10,5))
cagr_series.plot(kind="bar")

plt.title("Population CAGR by State (1981–2025)")
plt.ylabel("CAGR (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.heatmap(df5.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation of ERP Data')
plt.show()
```

```python
df5_clean = df5.drop(columns = ['Net Permanent and Long Term Movement','Migration Adjustment '])
df5_clean.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df5_clean.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation of ERP Data')
plt.show()
```

```python
components = df5_clean[['Natural Increase ', 'Net Overseas Migration ']].resample('Y').mean()
components['total_growth'] = components.sum(axis=1)
components.plot(title="Demand Decomposition: Natural vs Migration", figsize=(10,6))
```

```python
df5_clean.columns
```

```python
df5_annual = df5_clean.resample('Y').sum()

plt.figure(figsize=(12,6))

plt.plot(df5_annual.index, 
         (df5_annual['Overseas Arrivals']) - 
         (df5_annual['Overseas Departures ']),
         color='red', linewidth=2.5, label='Net Migration')
plt.title('Arrivals vs Departures: Net Population Gain')
plt.legend()
plt.ylabel('Number of People')
plt.show()
```

```python

components = df5_annual[['Natural Increase ', 'Net Overseas Migration ']].copy()
plot_components =components[['Natural Increase ', 'Net Overseas Migration ']].resample('Y').sum()
components['Total Growth'] = components.sum(axis=1)


plot_components.index = plot_components.index.strftime('%Y')

# Plotting
ax = plot_components.plot(kind='bar', stacked=True, figsize=(15, 7), width=0.8)

plt.title('Annual ERP Growth: Natural Increase vs Net Overseas Migration')
plt.ylabel('People Added')
plt.xlabel('Year')
plt.legend(['Natural Increase (Births - Deaths)', 'Net Overseas Migration'])

# Optional: Rotate labels to be horizontal if they fit, or keep them vertical
plt.xticks(rotation=90) 
plt.show()
```

```python
df5_clean['Total Growth'] = df5_clean['Natural Increase '] + df5_clean['Net Overseas Migration ']
rolling = df5_clean.rolling(window=12).sum()

rolling['% Natural'] = (rolling['Natural Increase '] / rolling['Total Growth']) * 100
rolling['% Migration'] = (rolling['Net Overseas Migration '] / rolling['Total Growth']) * 100

rolling[['% Natural', '% Migration']].plot(figsize=(12,6), title='Rolling 12-Month Contribution to ERP Growth')
plt.ylabel('Percentage of Total Growth')
plt.show()
```
